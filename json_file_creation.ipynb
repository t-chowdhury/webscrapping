```python
### Coromant

def is_number(s):
    try:
        #print('calling is_number method:')
        s_corr = s.replace(',','.')
        if(complex(s_corr)):
            digits = s_corr.split('.')                
            if(len(digits) > 1):
                return 'real'
            else:
                return 'int'
    except ValueError:
        return 'String'
    
    
json_file_count = 0
for file in files_list:#[count-1:count]:    
    
    with open(file, encoding='utf8') as infile:
        product_json_dict = {}
        product_info_dict = {}
        product_json_dict = json.load(infile)    
        
        json_file_count = json_file_count + 1
        
        if((json_file_count%5000) == 0):
            print("Counting....................#", json_file_count)         
        
        product_info_dict['Id']          = product_json_dict['Id']
        if(product_json_dict['Name'] == ''):
            product_info_dict['Name']        = None
        else:
            product_info_dict['Name']        = product_json_dict['Name']
            
        product_info_dict['Description'] = product_json_dict['Description']
        product_info_dict['Group']       = None
        product_info_dict['Grade']       = None
        product_info_dict['Structure']   = [product_json_dict['Structure']]
        
        attr_str = product_json_dict['Attributes']

        product_values = []
        product_attributes = []
        add_value = ''
        
        for line in attr_str.split('\n'):
            flag = 1
            if ( (line.endswith(')')) & ('CoroMill 300' not in line) ):
                #print(line)
                product_attributes.append(line)
                flag=0  

                if(add_value != ''):
                    product_values.append(add_value) 
                    add_value = ''
            else:        
                if(flag == 1):            
                    add_value = line + ',' + add_value  

        ## add the values of last column
        product_values.append(add_value)
        
        product_attr_list = []
        for attr, value in zip(product_attributes, product_values):
            product_unit = None
            elem = re.split(r'[()]',attr)    

            if( (len(elem[-1]) == 0) & (elem[-2].isupper()) ):
                product_description = ''.join(elem[0:-2])
                product_name = elem[-2]
            
            
            product_value = value.strip(',')
            
            if('mm' in value):
                product_value = value.strip(',').split('mm')[0].strip(' ')
                product_unit  = 'mm'
                
            if('kg' in value):
                product_value = value.strip(',').split('kg')[0].strip(' ')
                product_unit  = 'kg'                
                
            if('bar' in value):
                product_value = value.strip(',').split('bar')[0].strip(' ')
                product_unit  = 'bar'                

            if('deg' in value):
                product_value = value.strip(',').split('deg')[0].strip(' ')
                product_unit  = 'deg'           
                
            if('r/min' in value):
                product_value = value.strip(',').split('r/min')[0].strip(' ')
                product_unit  = 'r/min'     
                
            if('Nm' in value):
                product_value = value.strip(',').split('Nm')[0].strip(' ')
                product_unit  = 'Nm'                 
                

            #print(product_description, '->', product_name, '->', product_value, '->', product_unit)              
            product_attr_dict = {}
            
            product_attr_dict['Name']        = product_name
            product_attr_dict['DataType']    = is_number(product_value)
            product_attr_dict['Description'] = product_description.strip(' ')
            product_attr_dict['Value']       = product_value
            product_attr_dict['Unit']        = product_unit
            
            product_attr_list.append(product_attr_dict)
            
        product_info_dict['Attributes']      = product_attr_list             
            
        with open('Coromant.txt', 'a', encoding='utf-8') as outfile:
            json.dump(product_info_dict, outfile, sort_keys=False, ensure_ascii=False)
            outfile.write("\n")

print("Number of json files #", json_file_count)
```
